```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  reading the data
df = pd.read_csv("amazon.csv")
```

```python
# convert columns discounted_price,actual_price,discount_percentage,rating,rating_count to the numerical column

# Remove the '₹' symbol and commas from the columns
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')

# Convert the columns to numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%', ''), errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')
```

```python
df.info()
```

```python
df['category'].head()
```

```python
#  transform category  column and keep only first  word before separator '|"

df['category'] = df['category'].str.split('|').str[0]
```

# Exploratory Data Analysis

```python
# Question : What is the average rating for each product category

avg_rating_by_category = df.groupby('category')['rating'].mean()
print(avg_rating_by_category)
```

```python
print("Insights: ")

# Find the categories with the highest average ratings.
top_rated_categories = avg_rating_by_category.sort_values(ascending=False).head(10)
print("Top 10 categories with the highest average ratings:\n", top_rated_categories)

# Find the categories with the lowest average ratings.
bottom_rated_categories = avg_rating_by_category.sort_values(ascending=True).head(10)
print("\nTop 10 categories with the lowest average ratings:\n", bottom_rated_categories)

# Visualize the average rating by category.
plt.figure(figsize=(5, 3))
sns.barplot(x=avg_rating_by_category.values, y=avg_rating_by_category.index)
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
# Question: What are the top rating_count products by category?

# Group by category and find the product with the highest rating count in each category
top_rating_count_by_category = df.loc[df.groupby('category')['rating_count'].idxmax()]

# Display the top rating_count products by category
print(top_rating_count_by_category[['category', 'product_name', 'rating_count']])

# Visualize the top rating_count products by category
plt.figure(figsize=(10, 6))
sns.barplot(x='rating_count', y='product_name', data=top_rating_count_by_category, hue='category')
plt.title('Top Rating Count Products by Category')
plt.xlabel('Rating Count')
plt.ylabel('Category')
plt.show()

# Insights
print("Insights:")
print("This graph shows the top rating count products for each category. It helps identify the most popular products within each category based on customer reviews.")
print("We can see that some categories have significantly higher rating counts than others, indicating greater customer engagement and interest in those products.")
print("This information can be valuable for understanding customer preferences and for making decisions related to marketing, inventory, and product development.")
```

```python
# Question: What is the distribution of discounted prices vs. actual prices?

# Visualize the distribution of discounted prices vs. actual prices
plt.figure(figsize=(10, 6))
sns.histplot(df['discounted_price'], bins=50, kde=True, color='blue', label='Discounted Price')
sns.histplot(df['actual_price'], bins=50, kde=True, color='red', label='Actual Price')
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Insights
print("Insights:")
print("This histogram shows the distribution of discounted prices and actual prices.")
print("We can observe the overall price range and the frequency of products within different price brackets.")
print("This helps understand the pricing strategy and the distribution of products across different price points.")
```

```python
# Question: How does the average discount percentage vary across categories?

# Calculate the average discount percentage for each category
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean()

# Print the results
print(avg_discount_by_category)

# Visualize the average discount percentage by category
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_discount_by_category.values, y=avg_discount_by_category.index)
plt.title('Average Discount Percentage by Category')
plt.xlabel('Average Discount Percentage')
plt.ylabel('Category')
plt.show()
```

**Insights**:

This graph shows the average discount percentage for each category.
We can see that some categories have significantly higher average discounts than others.
This could indicate that certain categories are more competitive or have higher demand, leading to more frequent discounts.
Understanding these variations can help businesses optimize their pricing strategies and promotions.

```python
# Question: What are the most popular product names?

# Group by product name and count the occurrences
product_counts = df['product_name'].value_counts().head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=product_counts.values, y=product_counts.index)
plt.title('Top 10 Most Popular Product Names')
plt.xlabel('Count')
plt.ylabel('Product Name')
plt.show()
```

```python
# Question: What are the most popular product keywords?

from collections import Counter

# Combine all product names into a single string
all_product_names = " ".join(df['product_name'].astype(str).tolist())

# Split the string into individual words
words = all_product_names.lower().split()

# Remove common words (stop words)
stop_words = ["for", "with", "and", "the", "a", "in", "on", "of", "to", "is", "it", "that", "this", "was", "are", "were", "be", "been", "being","|","&","-"]
words = [word for word in words if word not in stop_words]

# Count the frequency of each word
word_counts = Counter(words)

# Get the most common words
most_common_words = word_counts.most_common(10)

# Print the results
print(most_common_words)

#generate graph
sns.barplot(x=[count for word, count in most_common_words], y=[word for word, count in most_common_words])
plt.title('Most Popular Product Keywords')
plt.xlabel('Count')
plt.ylabel('Keyword')
plt.show()
```

**Insights:**

From aboveplot we can see that most used keywords are usb,cable,charging,fast,C,smart etc

```python
# Question: What is the correlation between discounted_price and rating?

correlation = df['discounted_price'].corr(df['rating'])
print(f"Correlation between discounted_price and rating: {correlation}")
```

```python
# Question: What are the Top 5 categories based on the highest ratings?

# Calculate the average rating for each category
avg_rating_by_category = df.groupby('category')['rating'].mean()

# Sort the categories by average rating in descending order
top_categories = avg_rating_by_category.sort_values(ascending=False)

# Get the top 5 categories
print(top_categories.head(5))


#generate graph
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title('Top 5 Categories based on Highest Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
# Question: Identify any potential areas for improvement or optimization based on the data analysis.

# Analyze the distribution of rating counts.
plt.figure(figsize=(10, 6))
sns.histplot(df['rating_count'], bins=50, kde=True)
plt.title('Distribution of Rating Counts')
plt.xlabel('Rating Count')
plt.ylabel('Frequency')
plt.show()

# Insights:
print("Insights:")
print("This histogram shows the distribution of rating counts. We can observe if there are any outliers or if the data is skewed.")
print("Potential areas for improvement:")
print("- If there are a significant number of products with very low rating counts, it might indicate a need to encourage more customer reviews.")
print("- If the distribution is skewed, it might suggest that certain products are dominating the rating counts, while others are neglected.")

# Analyze the relationship between discount percentage and rating.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title('Relationship between Discount Percentage and Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.show()

# Insights:
print("Insights:")
print("This scatter plot shows the relationship between discount percentage and rating. We can observe if there is any correlation between the two variables.")
print("Potential areas for improvement:")
print("- If there is a negative correlation, it might suggest that heavily discounted products tend to have lower ratings.")
print("- If there is no clear correlation, it might indicate that discount percentage is not a major factor influencing customer ratings.")

# Analyze the distribution of actual prices.
plt.figure(figsize=(10, 6))
sns.histplot(df['actual_price'], bins=50, kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()

# Insights:
print("Insights:")
print("This histogram shows the distribution of actual prices. We can observe if there are any outliers or if the data is skewed.")
print("Potential areas for improvement:")
print("- If there are a significant number of products with very high prices, it might indicate a need to adjust pricing strategies for those products.")
print("- If the distribution is skewed, it might suggest that the majority of products are concentrated in a specific price range.")
```
